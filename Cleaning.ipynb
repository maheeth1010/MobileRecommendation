## To get techcenturion score of Processors

```python
import requests 
from bs4 import BeautifulSoup 
import csv

url = "https://www.techcenturion.com/smartphone-processors-ranking"
r = requests.get(url)
soup = BeautifulSoup(r.content, 'html5lib')
trs = soup.find_all('tr')
cpu = {}
for tr in trs:
    tds = tr.find_all('td')
    name = tds[1].get_text().strip()
    if '+' in name:
        name = name.strip('+')
        name += " Plus"
    if 'Apple' in name:
        name = name.strip('Apple ')
        name = 'A'+name
    score = tds[3].get_text().strip('*')
    cpu[name] = score
```

## Nanoreview score of Processors

```python
import requests 
from bs4 import BeautifulSoup 
import csv

url = "https://nanoreview.net/en/soc-list/rating"
r = requests.get(url)
soup = BeautifulSoup(r.content, 'html5lib')
links = soup.find_all('tr')
links = links[1:]
```

```python
antutu = {}
for i in links:
    tds = i.find_all('td')
    name = tds[1].find('a').get_text()
    score = tds[3].get_text()
    if '-' in score:
        antutu[name] = '0'
    else:
        antutu[name] = score
#     print(tds[1])
#     print(tds[3])
#     print()
#     print()
#     print()
#     print()
```

```python
antutu
```

```python
cpus = []
cpuScore = []
count = 0
for i in df['CPU']:
    comma = i.split(',')
    i = comma[0]
    if count==8 or count==14 or count==15:
        cpus.append(i)
        cpuScore.append('0')
        count = count + 1
        continue
    if '+' in i:
        ans = i.split('+')
        i = ans[0]+" Plus "+ans[1]
        placed = False
    for k,v in antutu.items():
        spaced = k.split(' ')
        check = False
        for j in spaced:
            if j not in i:
                check = True
                break
        if check == False:
            cpus.append(k)
            cpuScore.append(v)
            print(str(count)+" "+k)
            placed = True
            break
    if placed == False:
        cpus.append(i)
        cpuScore.append('0')
    count = count + 1
```
